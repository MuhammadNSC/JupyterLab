Data Aggregation
Group the coffee production data by type for each year from 1990 to 2020.
Summarize the data to reflect total production volumes for each coffee type per year.

```python
# Connect SQLite database 
conn = sqlite3.connect(':memory:') 

# SQL script to connect the SQLite database
conn.executescript(sql_script)

# SQL query to retrieve the coffee_production data from 1990 to 2020
sql_query = """
    SELECT coffee_type,
       SUM("1990_1991") AS "1990_1991",
       SUM("1991_1992") AS "1991_1992",
       SUM("1992_1993") AS "1992_1993",
       SUM("1993_1994") AS "1993_1994",
       SUM("1994_1995") AS "1994_1995",
       SUM("1995_1996") AS "1995_1996",
       SUM("1996_1997") AS "1996_1997",
       SUM("1997_1998") AS "1997_1998",
       SUM("1998_1999") AS "1998_1999",
       SUM("1999_2000") AS "1999_2000",
       SUM("2000_2001") AS "2000_2001",
       SUM("2001_2002") AS "2001_2002",
       SUM("2002_2003") AS "2002_2003",
       SUM("2003_2004") AS "2003_2004",
       SUM("2004_2005") AS "2004_2005",
       SUM("2005_2006") AS "2005_2006",
       SUM("2006_2007") AS "2006_2007",
       SUM("2007_2008") AS "2007_2008",
       SUM("2008_2009") AS "2008_2009",
       SUM("2009_2010") AS "2009_2010",
       SUM("2010_2011") AS "2010_2011",
       SUM("2011_2012") AS "2011_2012",
       SUM("2012_2013") AS "2012_2013",
       SUM("2013_2014") AS "2013_2014",
       SUM("2014_2015") AS "2014_2015",
       SUM("2015_2016") AS "2015_2016",
       SUM("2016_2017") AS "2016_2017",
       SUM("2017_2018") AS "2017_2018",
       SUM("2018_2019") AS "2018_2019",
       SUM("2019_2020") AS "2019_2020"
FROM coffee_production
GROUP BY coffee_type;
"""

# results of the SQL query into a pandas DataFrame
aggregated_data = pd.read_sql_query(sql_query, conn)

# Close the database connection
conn.close()

# Display the aggregated data
aggregated_data.head()
```

Data Transformation
Transpose the aggregated data so that coffee types become the rows and years become the columns.
This reorientation will prepare your data for the next step of correlation analysis.

```python
# Transpose the aggregated data
transposed_data = aggregated_data.set_index('coffee_type').transpose()

# Display the transposed data
transposed_data.head()
```

Correlation Analysis
Using the Pandas library in Python, apply the corr() method to your transposed dataset to compute the correlation matrix.
The correlation matrix should show the relationship coefficients between each pair of coffee types.

```python
# Compute the correlation matrix
correlation_matrix = transposed_data.corr()

# Display the correlation matrix
correlation_matrix.head()
```
